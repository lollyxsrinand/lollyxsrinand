```python
from collections import defaultdict
import math
documents = [
    ("AI is awesome", "Tech"),
    ("AI is future", "Tech"),
    ("Computing is great", "Tech"),
    ("Weather is nice", "Non-Tech"),
    ("Food is tasty", "Non-Tech"),
    ("Movie is fun", "Non-Tech")
]

text_sentence = input("Enter text")
def preprocess(text):
    return text.lower().split()

word_counts = defaultdict(lambda: defaultdict(int))
class_counts = defaultdict(int)
total_words = defaultdict(int)
vocab = set()
for text, label in documents:
    class_counts[label] += 1
    words = preprocess(text)

    for word in words:
        word_counts[label][word] += 1
        total_words[label] += 1
        vocab.add(word)

V = len(vocab)
total_docs = len(documents)

def classify(sentence):
    words = preprocess(sentence)
    scores = {}

    scores[label] = math.log(class_counts[label]/total_docs)

    for word in words:
        scores[label] += math.log(
            (word_counts[label][word] + 1) / (total_words[label] + V)
        )

    return max(scores, key=scores.get)
predicted_class = classify(text_sentence)
print("Test sentence", text_sentence)
print("predicted class", predicted_class)
```
