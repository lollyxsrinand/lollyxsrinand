```python
# week 3
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

documents = ['ML is a part of AI',
             'Deep Learning uses NN',
             'AI and ML are related',
             'football is popular outdoor sport',
]

vectorizer = TfidfVectorizer(stop_words='english')
x = vectorizer.fit_transform(documents)

z_cosine = linkage(x.toarray(), method='average', metric='cosine')
plt.figure(figsize = (8, 5))
dendrogram(z_cosine, labels=documents)
plt.title('cosine distance')
plt.xticks(rotation = 45)
plt.show()

z_euclidean = linkage(x.toarray(), method='average', metric='euclidean')
plt.figure(figsize = (8, 5))
dendrogram(z_euclidean, labels = documents)
plt.title("euclidean distance")
plt.xticks(rotation = 45)
plt.show()

z_manhattan = linkage(x.toarray(), method="average", metric="cityblock")
plt.figure(figsize = (8, 5))
dendrogram(z_manhattan, labels = documents)
plt.title("manhattan distance")
plt.xticks(rotation = 45)
plt.show()

clusters = fcluster(z_cosine, criterion='maxclust', t=2)
print("clusters:", clusters)
```
