```python
# week 4
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean, cityblock

documents = ['web data mining techniques are useful for info retrieval',
             'info retrieval uses data mining and web mining methods',
             'machine learning techniques are used in webapps']

Vectorizer = TfidfVectorizer()

tfidf_matrix = Vectorizer.fit_transform(documents)
tfidf_dense = tfidf_matrix.toarray()

scaler = StandardScaler()
Standardized_matrix = scaler.fit_transform(tfidf_dense)
print("Standardized document vector\n", Standardized_matrix)

def eucledian_distance(doc1, doc2):
    return euclidean(doc1, doc2)

def manhattan_dist(doc1, doc2):
    return cityblock(doc1, doc2)

def cosine_distance(doc1, doc2):
    similarity = cosine_similarity([doc1],[doc2])[0][0]
    return 1 - similarity

doc1 = Standardized_matrix[0]
doc2 = Standardized_matrix[1]

print("eucledian distance", eucledian_distance(doc1, doc2))
print("manhattan distance", manhattan_dist(doc1, doc2))
print("cosine distance", cosine_distance(doc1, doc2))
```
