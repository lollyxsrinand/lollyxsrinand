```python
# week 5
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.util import ngrams
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer

text = """ Text Mining is the process of extracting useful information from large amounts of text data text preprecessing is an improtant zxtep in natural language proddcessing"""

tokens = word_tokenize(text.lower())
print("Tokens:", tokens)

stop_words = set(stopwords.words("english"))
filtered_words = [word for word in tokens if word.isalnum() and word not in stop_words]
print("\n filtered wrods:", filtered_words)

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print('\n stemmed wrods', stemmed_words)

bigrams = list(ngrams(stemmed_words, 2))
bigram_freq = Counter(bigrams)
print('\n Bigram frequency', bigram_freq)

trigrams = list(ngrams(stemmed_words, 3))
trigram_freq = Counter(trigrams)
print('\ntrigram frequencey', trigram_freq)

vectorize_bigram = CountVectorizer(stop_words='english', ngram_range=(2, 2))
bigram_matrix = vectorize_bigram.fit_transform([text])
print('\n bigram features', vectorize_bigram.get_feature_names_out())

vectorize_trigram = CountVectorizer(stop_words='english', ngram_range=(3, 3))
trigram_matrix = vectorize_trigram.fit_transform([text])
print('\nTrigram features:')
print(vectorize_trigram.get_feature_names_out())
```
